# 12. Integer to Roman

Seven different symbols represent Roman numerals with the following values:

| Symbol | Value |
| :----: | ----: |
|    I   |     1 |
|    V   |     5 |
|    X   |    10 |
|    L   |    50 |
|    C   |   100 |
|    D   |   500 |
|    M   |  1000 |

Roman numerals are formed by appending the conversions of decimal place values from highest to lowest. Converting a decimal place value into a Roman numeral has the following rules:

* If the value does not start with 4 or 9, select the symbol of the maximal value that can be subtracted from the input, append that symbol to the result, subtract its value, and convert the remainder to a Roman numeral.
* If the value starts with 4 or 9 use the **subtractive form** representing one symbol subtracted from the following symbol, for example, 4 is 1 (`I`) less than 5 (`V`): `IV` and 9 is 1 (`I`) less than 10 (`X`): `IX`. Only the following subtractive forms are used: 4 (`IV`), 9 (`IX`), 40 (`XL`), 90 (`XC`), 400 (`CD`) and 900 (`CM`).
* Only powers of 10 (`I`, `X`, `C`, `M`) can be appended consecutively at most 3 times to represent multiples of 10. You cannot append 5 (`V`), 50 (`L`), or 500 (`D`) multiple times. If you need to append a symbol 4 times use the **subtractive form**.

Given an integer, convert it to a Roman numeral.

---

### Example 1:

**Input:** `num = 3749`
**Output:** `"MMMDCCXLIX"`

**Explanation:**

```
3000 = MMM as 1000 (M) + 1000 (M) + 1000 (M)
 700 = DCC as 500 (D) + 100 (C) + 100 (C)
  40 = XL as 10 (X) less of 50 (L)
   9 = IX as 1 (I) less of 10 (X)
Note: 49 is not 1 (I) less of 50 (L) because the conversion is based on decimal places
```

### Example 2:

**Output:** `"LVIII"`

**Explanation:**

```
50 = L
 8 = VIII
```

### Example 3:

**Input:** `num = 1994`
**Output:** `"MCMXCIV"`

**Explanation:**

```
1000 = M
 900 = CM
  90 = XC
   4 = IV
```

### Constraints

* `1 <= num <= 3999`


````python
def integer_to_roman_numeral(y:int)-> str:
    blerb = str(y)
    digit_count = len(blerb)
    fin_num = ""
    for index, digit in enumerate(blerb):
        cur_dig = digit_count - index
        if digit == "0":
            continue
        if cur_dig == 4:
            one = "M"
            fin_num += one * int(digit)
        elif cur_dig == 3:
            one = "C"
            five = "D"
            nex_one = "M"
            if digit <= "3":
                fin_num += one * int(digit)
            elif digit == "4":
                fin_num += one + five
            elif digit == "9":
                fin_num += one + nex_one
            else:
                fin_num += five + one * (int(digit)-5)
        elif cur_dig == 2:
            one = "X"
            five = "L"
            nex_one = "C"
            if digit <= "3":
                fin_num += one * int(digit)
            elif digit == "4":
                fin_num += one + five
            elif digit == "9":
                fin_num += one + nex_one
            else:
                fin_num += five + one * (int(digit)-5)
        else:
            one = "I"
            five = "V"
            nex_one = "X"
            if digit <= "3":
                fin_num += one * int(digit)
            elif digit == "4":
                fin_num += one + five
            elif digit == "9":
                fin_num += one + nex_one
            else:
                fin_num += five + one * (int(digit)-5)
    return fin_num
````

## 0ms Runtine Code Example

````python
class Solution:
    # Leetcode
    def intToRoman(self, num):
        val_to_roman = [
            (1000, 'M'),
            (900,  'CM'),
            (500,  'D'),
            (400,  'CD'),
            (100,  'C'),
            (90,   'XC'),
            (50,   'L'),
            (40,   'XL'),
            (10,   'X'),
            (9,    'IX'),
            (5,    'V'),
            (4,    'IV'),
            (1,    'I')
        ]

        roman = ""
        for value, symbol in val_to_roman:
            while num >= value:
                roman += symbol
                num -= value

        return roman
````

````python
class Solution:
    # Chat GPT 5

    def intToRoman(self, num: int) -> str:
        thousands = ["", "M", "MM", "MMM"]
        hundreds  = ["", "C", "CC", "CCC", "CD", "D", "DC", "DCC", "DCCC", "CM"]
        tens      = ["", "X", "XX", "XXX", "XL", "L", "LX", "LXX", "LXXX", "XC"]
        ones      = ["", "I", "II", "III", "IV", "V", "VI", "VII", "VIII", "IX"]

        return (
            thousands[num // 1000] +
            hundreds[(num // 100) % 10] +
            tens[(num // 10) % 10] +
            ones[num % 10]
        )
````

Chat GPT Explanation of the differences of why my 3ms runtine code, which was O(N) and is top 87.45% on Leetcode, is still not optimal

Here’s why my version tends to be slower:

1. **String ↔ int conversions**
   You do `int(digit)` multiple times per digit and compare strings like `digit <= "3"`—that’s extra parsing and branching the other code avoids.

2. **More branching in Python space**
   Your per-place `if/elif` ladders (thousands/hundreds/tens/ones) execute several Python bytecode ops each time. The accepted code just walks a fixed table with a tight `while` loop—fewer branches overall.

3. **Temporary variables every digit**
   Rebinding `one`, `five`, `nex_one` each iteration adds bytecode and dictionary lookups (for names) in the interpreter.

4. **Same concatenation cost, but more overhead around it**
   Both solutions build the result with `roman += ...` (which creates new strings). But since the greedy solution performs fewer additional Python-level operations around each concat, it often wins.

If you want the fastest Pythonic approach, precompute four small lookup tables and do **exactly four concatenations**—no loops, no branches, no conversions:

```python
class Solution:
    def intToRoman(self, num: int) -> str:
        thousands = ["", "M", "MM", "MMM"]
        hundreds  = ["", "C", "CC", "CCC", "CD", "D", "DC", "DCC", "DCCC", "CM"]
        tens      = ["", "X", "XX", "XXX", "XL", "L", "LX", "LXX", "LXXX", "XC"]
        ones      = ["", "I", "II", "III", "IV", "V", "VI", "VII", "VIII", "IX"]

        return (
            thousands[num // 1000] +
            hundreds[(num // 100) % 10] +
            tens[(num // 10) % 10] +
            ones[num % 10]
        )
```

Why this is fast:

* Constant-time arithmetic indexing (no `int()` conversions from chars).
* Zero Python loops/`while`s inside—just 4 table lookups and 4 concats.
* Minimal branching and bytecode.

If you prefer to keep your style, you can speed it up by:

* Converting `digit` once: `d = ord(digit) - 48` (since `'0'..'9'` are consecutive) and using `d` in all tests.
* Hoisting the per-place tables (like above) instead of `if/elif`.
* Avoiding repeated `int(digit)` calls.
